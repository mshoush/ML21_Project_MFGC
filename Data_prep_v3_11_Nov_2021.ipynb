# 1. read and rename data

```python
def read_and_rename_data(dirr):
    # use glob to get all the csv files
    # in the folder
    path = dirr
    csv_files = glob.glob(os.path.join(path, "*.xlsx"))


    # loop over the list of csv files
    for f in csv_files:
        # read the csv file
        print(f)

        df = pd.read_excel(f)
        df = df.rename(columns={'Registrikood': 'Registration_number',\
                                'Nimi': 'Name', \
                               'Liik': 'Type',\
                                'Registreeritud käibemaksukohustuslaste registrisse': 'Registered_for_VAT', \
                                'EMTAK tegevusvaldkond, mis on EMTAKi struktuuris tähistatud tähtkoodiga': 'Business_area_by_EMTAK_system',\
                                'Maakond': 'County', \
                                'Riiklikud Maksud': 'All_taxes',\
                                'Tööjõumaksud Ja Maksed': 'Labor_taxes', \
                                'Kaive': 'Turnover',\
                                'Tootajaid': 'number_of_employees'})
        year = str(f).split("/")[2].split("_")[2]
        quarter = str(f).split("/")[2].split("_")[3]
        df['year_quarter'] = str(year)+"_"+str(quarter)
  
        # Tootajaid
        
        #df = df.set_axis(["Registration_number", "Name", "Type", "Registered_for_VAT", \
 #"Business_area_by_EMTAK_system", "County", "All_taxes", "Labor_taxes", \
 #"Turnover", "number_of_employees"], axis=1, inplace=False)
        df.to_csv("./data/csv/"+str(f).split("/")[2]+".csv", index=False)
        #break
        
```

### 1. The whole period (2017 : 2021), including all quarters. 1st data frame.

```python
import os
import glob
import pandas as pd
os.chdir("./data/csv/")

extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
#export to csv
combined_csv.to_csv( "./final_csv_df/all_df_2017_2021.csv", index=False)
```

### 3. During Covid period  (2nd-4th-2020 : 2021). 3rd data frame.

```python
import os
import glob
import pandas as pd
#os.chdir("./data/csv/")
df_list = []
extension = 'csv'
all_filenames = [df_list.append(i) for i in glob.glob('*.{}'.format(extension))]
df_1st_2020 = df_list[4]
names_rest = df_list[7:-1]
list_of_before_covid_df = names_rest + [df_1st_2020]
list_of_during_covid_df = sorted(df_list)[15:]

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in list_of_before_covid_df ])
#export to csv
combined_csv.to_csv( "./final_csv_df/before_covid_1st2017_1st2020.csv", index=False)

#combine all files in the list
combined_csv = pd.concat([pd.read_csv(f) for f in list_of_during_covid_df ])
#export to csv
combined_csv.to_csv( "./final_csv_df/during_covid_2ndq2020_2021.csv", index=False)
```
