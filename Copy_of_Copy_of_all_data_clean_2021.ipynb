```python
import pandas as pd
import os
import glob
import numpy as np
import matplotlib.pyplot as plt

#1- replacing -ve values with very small number,
#2- For deletion date: we need to replace Nan with a valuable information. However, you need to think about this replacement since it should be consistent with others that are still exist. Maybe one possible solution is all companies that are exist will take 1, and others will take 0, instead of the exact date. I did it this way, existing companies are 1 and non existing 0
#3- Can you check if we have any outlier or inconsistent t data values at any column or row.
#4- use index=False with to_csv saving function.
#5- we have two columns with the name of the company, Name_x, and Name_y. please drop one of them.
```

```python
df = pd.read_csv('During_Covid_cleaned.csv')
df.head()
```

```python
#filling missing values with 1 
df['Deletion_date']=df['Deletion_date'].fillna(1)
df
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['Name_y'])
```

```python
df["County"].fillna("No County", inplace = True) 
```

```python
df.isnull().sum()
```

```python
df["Business_area_by_EMTAK_system"].fillna("Missing Business area", inplace = True) 
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['number_of_employees'], how = 'all', inplace=True)
```

```python
df.isnull().sum()
```

```python
#dropping name_y
df.drop(columns=['Name_y'], inplace=True)
```

```python
df.isnull().sum()
```

```python
#converting float to int
cols = ['number_of_employees']
df[cols]= df[cols].applymap(np.int64)
df
```

```python
df.info()
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df.info()
```

```python
#replacing negative turover with small nr 
df['Turnover'] = df['Turnover'].apply(lambda x : x if x > 0 else 0.000001)
df
```

```python
#replacing all deleted copanies with 0

df['Deletion_date'] = df['Deletion_date'].apply(lambda x : x if x == 1 else 0)
df
```

```python
df.to_csv('During_Covid_Cleaned2.csv', index=False)
```
